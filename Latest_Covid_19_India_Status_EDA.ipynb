```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Latest Covid-19 India Status.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().count()
```

```python
df.drop_duplicates(inplace=True)
```

```python
states = df['State/UTs']
TC = df['Total Cases']
plt.pie([TC],labels=states, radius=5)
plt.show()
```

```python
sns.pairplot(df, hue='State/UTs')
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
corr1 = df[['Active', 'Discharged']].corr()
sns.heatmap(corr1,annot=True)
```

```python
df1 = df.iloc[:,0:2]
df1 = df1.sort_values(by='Total Cases', ascending=False)
df1.head()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel('States and UTs')
plt.ylabel('Total Cases')
plt.bar(df1['State/UTs'],df1['Total Cases'])
plt.show()
```

```python
df2 = df[['State/UTs','Active']]
df2 = df2.sort_values(by='Active', ascending=False)
df2.head()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel('States and UTs')
plt.ylabel('Active')
plt.bar(df2['State/UTs'],df2['Active'])
plt.show()
```

```python
corr2 = df[['Total Cases', 'Active', 'Discharged', 'Deaths']].corr()
sns.heatmap(corr2,annot=True)
```

```python
assam=df.iloc[3,:]
arp=df.iloc[2,:]
manipur=df.iloc[21,:]
meghalaya=df.iloc[22,:]
mizoram=df.iloc[23,:]
ngl=df.iloc[24,:]
```

```python
plt.title('Comparison of the total cases in each of the seven sister states of India')
plt.bar(['Arunachal Pradesh','Assam','Manipur','Meghalaya','Mizoram','Nagaland'],[arp['Total Cases'],assam['Total Cases'],manipur['Total Cases'],meghalaya['Total Cases'],mizoram['Total Cases'],ngl['Total Cases']])
plt.xlabel('States')
plt.ylabel('Total Cases')
plt.show()
```

```python
plt.title('Comparison of the active cases in each of the seven sister states of India')
plt.bar(['Arunachal Pradesh','Assam','Manipur','Meghalaya','Mizoram','Nagaland'],[arp['Active'],assam['Active'],manipur['Active'],meghalaya['Active'],mizoram['Active'],ngl['Active']])
plt.xlabel('States')
plt.ylabel('Active')
plt.show()
```

```python
plt.title('Comparison of the discharged cases in each of the seven sister states of India')
plt.bar(['Arunachal Pradesh','Assam','Manipur','Meghalaya','Mizoram','Nagaland'],[arp['Discharged'],assam['Discharged'],manipur['Discharged'],meghalaya['Discharged'],mizoram['Discharged'],ngl['Discharged']])
plt.xlabel('States')
plt.ylabel('Discharged')
plt.show()
```

```python
plt.title('Comparison of the deaths in each of the seven sister states of India')
plt.bar(['Arunachal Pradesh','Assam','Manipur','Meghalaya','Mizoram','Nagaland'],[arp['Deaths'],assam['Deaths'],manipur['Deaths'],meghalaya['Deaths'],mizoram['Deaths'],ngl['Deaths']])
plt.xlabel('States')
plt.ylabel('Deaths')
plt.show()
```
